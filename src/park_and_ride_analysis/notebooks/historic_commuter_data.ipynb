```python
import matplotlib.pyplot as plt

# Data for the commuters coming to Darmstadt city from other cities
years = [2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023]
commuters = [64188, 65751, 66098, 67990, 69744, 71233, 72603, 72819, 74867, 76492, 76437]

# Create a plot
plt.figure(figsize=(14, 6))

# Line diagram
# plt.subplot(1, 2, 1)
plt.plot(years, commuters, marker='o', linestyle='-', color='b')
plt.xlabel('Year')
plt.ylabel('Number of Commuters')
# plt.title('Line Diagram: Number of Commuters Coming to Darmstadt City (2013-2023)')
plt.grid(True)
plt.xticks(years, rotation=45)

# Bar diagram
# plt.subplot(1, 2, 2)
# plt.bar(years, commuters, color='g')
# plt.xlabel('Year')
# plt.ylabel('Number of Commuters')
# plt.title('Bar Diagram: Number of Commuters Coming to Darmstadt City (2013-2023)')
# plt.xticks(years, rotation=45)

plt.tight_layout()
plt.show()
```
